```python
import numpy as np
import yt
import matplotlib.pylab as plt
import matplotlib
```

```python
#dss= [ yt.load(filename) for filename in [
#    "~/source/GP3/build/plt00000","~/source/GP3/build/plt00100","~/source/GP3/build/plt00500","~/source/GP3/build/plt01000"]
#    ]

files=[ "~/source/GP3/build/plt{:05d}/".format(i) for i in range(0,1000,200) ]
dss= [ yt.load(filename) for filename in files ] 
```

```python
%matplotlib notebook
i=0
cmap=matplotlib.cm.get_cmap('viridis')
for ds in dss:
    i+=1
    for g in ds.index.grids:
        colors=["red","blue"]
        r2=np.sqrt(g["x"]**2 + g["y"]**2 + g["z"]**2)
        #r2I=np.sqrt(gI["x"]**2 + gI["y"]**2 + gI["z"]**2)
        marker = "o" if g.Level ==0 else "^"
        color=cmap(i/len(dss))
        plt.scatter(r2.flatten(),g["phi"].flatten(),marker=marker,color=color,alpha=0.1)
        #plt.scatter(r2I.flatten(),gI["phi"].flatten(),color=colors[g.Level],
        #            facecolors='none',lw=1,alpha=0.1,marker="^")
plt.legend()
```
